```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = "/home/augustine77/Downloads/CIC/Wifi_and_MQTT/attacks/CSV/train/ARP_Spoofing_train.pcap.csv"
df = pd.read_csv(file)

# Basic descriptive statistics
print("Descriptive Statistics:\n", df.describe())

# Handling potential errors for non-numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# Distribution plots for numeric columns
num_plots = len(numeric_cols)
num_cols = 3  # Number of columns in the subplot grid
num_rows = (num_plots + num_cols - 1) // num_cols  # Number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easier iteration

for i, col in enumerate(numeric_cols):
    ax = axes[i]
    sns.histplot(df[col], ax=ax, kde=True)  # Histogram with KDE
    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()


# Analysis and further visualizations (example for a few columns):
print("\n--- Analysis and Further Visualizations ---\n")

# Example: Box plot for 'Duration' to check for outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Duration'])
plt.title('Box Plot of Duration')
plt.show()

# Example: Scatter plot to explore relationship between two variables (e.g., 'Rate' and 'Duration')
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Rate', y='Duration', data=df)
plt.title('Scatter Plot of Rate vs. Duration')
plt.show()

# Example: Pairplot for a subset of columns to see correlations
subset_cols = ['Rate', 'Duration', 'Srate', 'Drate']
sns.pairplot(df[subset_cols])
plt.show()

# Add more analysis and visualizations as needed based on your data and research questions.

```
